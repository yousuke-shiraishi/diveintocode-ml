### 学習用データと検証用データの分割

```python
from sklearn.preprocessing import StandardScaler
standardizer = StandardScaler()
#標準化
X = standardizer.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=66)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
print("--------------------------")

for n_neighbors in range(1,10,2):
    neigh_n = KNeighborsClassifier(n_neighbors=n_neighbors)
    neigh_n.fit(X_train, y_train)
    print(str(n_neighbors)+ "-nn")
    print("訓練スコア" + str(neigh_n.score(X_train,y_train)))
    pred_y = neigh_n.predict(X_test)
    print("実際のスコア" + str(neigh_n.score(X_test, y_test)))
    print("正答率" + str(accuracy_score(y_test, pred_y)))
    print("精度"+ str(precision_score(y_test, pred_y,average=None)))
    print("検出率" + str(recall_score(y_test, pred_y,average=None)))
    print("F値" + str( f1_score(y_test, pred_y,average=None)))
    print("混同マトリックス" + str(confusion_matrix(y_test,pred_y,labels=[1, 0])))
    print("---------------------------------------------------------------")#average='micro'を使いと同じ正答率も精度も検出率も同じになるのだがどういう事だ
```

sklearn.metrics.accuracy_scoreとは正解率を出すもの
正解率 (Accuracy)
全体から正解したものの割合
Accuracy = (TP + TN) / (TP + TN + FP + FN)
例えばTNが高いととりあえずNと判断すれば良いという判断になることもあるので注意が必要。

 sklearn.metrics.precision_score
 精度 (Precision)
 陽性と判断されたものの中で実際に陽性であったもの
 Precision = TP / (TP + FP)
 レコメンドで本人にあったものをお勧めするときなどに重要視される。
 確実に陽性であるというものしか選ばなければ精度は上がるが陽性なのに陰性であるものが取りこぼされる率が上がるので検出率が下がる

sklearn.metrics.recall_score
検出率 (Recall)
実際に陽性であるものを陽性であると判断される率
Recall = TPR = Sensitivity = TP / (TP + FN)
ガンなど判定の正確性を求められる場合に使われる。
ミスすることが許されない場合に重要視される。
なんでも陽性として判断しておけばこの率は上がるが反面、陰性なのに陽性と判断する率も上がるので精度が下がる

sklearn.metrics.f1_score
F 値 (F-measure, F-score, F1 Score とも呼ばれる) とは、精度 (Precision) と検出率 (Recall) をバランス良く持ち合わせているかを示す指標
F1 = 2 * (precision * recall) / (precision + recall)
0 〜 1 の間の数値で出力され、0 の場合最も悪い評価、1 の場合最も良い評価となる。

# 【問題6】可視化

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
neigh_n = KNeighborsClassifier(n_neighbors=3)
iris= load_iris()
X = iris.data
X = X[50:]
y =  iris.target[50:]
standardizer = StandardScaler()
X = standardizer.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=66)
X_train = X_train[:,[0,2]]
neigh_n.fit(X_train, y_train)
```

```python
def decision_region(X_train, y_train, model, step=0.01, title='decision region', xlabel='xlabel', ylabel='ylabel', target_names=['versicolor', 'virginica']):
    """
    2値分類を2次元の特徴量で学習したモデルの決定領域を描く。
    背景の色が学習したモデルによる推定値から描画される。
    散布図の点は学習用データである。

    Parameters
    ----------------
    X_train : ndarray, shape(n_samples, 2)
        学習用データの特徴量
    y_train : ndarray, shape(n_samples,)
        学習用データの正解値
    model : object
        学習したモデルのインスンタスを入れる
    step : float, (default : 0.1)
        推定値を計算する間隔を設定する
    title : str
        グラフのタイトルの文章を与える
    xlabel, ylabel : str
        軸ラベルの文章を与える
    target_names= : list of str
        凡例の一覧を与える
    """
    # setting
    scatter_color = ['red', 'blue']
    contourf_color = ['pink', 'skyblue']
    n_class = 2

    # pred
    mesh_f0, mesh_f1  = np.meshgrid(np.arange(np.min(X_train[:,0])-0.5, np.max(X_train[:,0])+0.5, step), np.arange(np.min(X_train[:,1])-0.5, np.max(X_train[:,1])+0.5, step))
    mesh = np.c_[np.ravel(mesh_f0),np.ravel(mesh_f1)]
    pred = model.predict(mesh).reshape(mesh_f0.shape)

    # plot
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.contourf(mesh_f0, mesh_f1, pred, n_class-1, cmap=ListedColormap(contourf_color))
    plt.contour(mesh_f0, mesh_f1, pred, n_class-1, colors='y', linewidths=3, alpha=0.5)
    for i, target in enumerate(set(y_train)):
        plt.scatter(X_train[y_train==target][:, 0], X_train[y_train==target][:, 1], s=80, color=scatter_color[i], label=target_names[i], marker='o')
    patches = [mpatches.Patch(color=scatter_color[i], label=target_names[i]) for i in range(n_class)]
    plt.legend(handles=patches)
    plt.legend()
    plt.show()
```

```python
decision_region(X_train,y_train,neigh_n )
```

# 【問題7】他の手法の学習

```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
iris= load_iris()
X = iris.data
X = X[50:]
y =  iris.target[50:]

standardizer = StandardScaler()
X = standardizer.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=66)
X_train = X_train[:,[0,2]]
X_test = X_test[:,[0,2]]
clfs =[]
neigh_n = KNeighborsClassifier(n_neighbors=3)
clfs.append(neigh_n)
#ロジスティック回帰
lg_clf = LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial')
clfs.append(lg_clf)
# SVM
svc_clf = SVC(gamma='auto')
clfs.append(svc_clf)
#決定木
dt_clf = DecisionTreeClassifier(random_state=0)
clfs.append(dt_clf)
#ランダムフォレスト 
rf_clf = RandomForestClassifier(n_estimators=100, max_depth=2,random_state=0)
clfs.append(rf_clf)
```

```python
cls_score=[]
cls_ac=[]
cls_prec=[]
cls_rec=[]
cls_f=[]
for clf in clfs:
    clf.fit(X_train,y_train)
    pred_y = clf.predict(X_test)
    cls_score.append(clf.score(X_test,y_test))
    cls_ac.append(accuracy_score(y_test, pred_y))
    cls_prec.append(precision_score(y_test, pred_y,average=None))
    cls_rec.append(recall_score(y_test, pred_y,average=None))
    cls_f.append(f1_score(y_test, pred_y,average=None))
    decision_region(X_train,y_train,clf)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
print("--------------------------")

for n_neighbors in range(1,10,2):
    neigh_n = KNeighborsClassifier(n_neighbors=n_neighbors)
    neigh_n.fit(X_train, y_train)
    print(str(n_neighbors) + "-nn")
    print("訓練スコア" + str(neigh_n.score(X_train,y_train)))
    pred_y = neigh_n.predict(X_test)
    print("実際のスコア" + str(neigh_n.score(X_test, y_test)))
    print("正答率" + str(accuracy_score(y_test, pred_y)))
    print("精度"+ str(precision_score(y_test, pred_y,average=None)))
    print("検出率" + str(recall_score(y_test, pred_y,average=None)))
    print("F値" + str( f1_score(y_test, pred_y,average=None)))
    print("混同マトリックス" + str(confusion_matrix(y_test,pred_y)))
    print("---------------------------------------------------------------")
```

```python
data = {"正答率":cls_ac,"実際のスコア":cls_score,"精度":cls_prec,"検出率":cls_rec,"f値":cls_f}
index = ["K近傍法","ロジスティック回帰","SVM","決定木","ランダムフォレスト"]
df = pd.DataFrame(data,index=index)
df
```
