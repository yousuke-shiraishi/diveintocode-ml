# 【課題1】データの取得

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris= load_iris()
```

```python
X = iris.data
y =  iris.target
df0 = pd.DataFrame(X, columns=iris.feature_names)
df0
```

```python
df1 = pd.DataFrame()
df1['Species'] = y
df1
```

# 【課題2】データの結合

load_iris}()はiris情報を全体に含んでいるのでこのオブジェクトを直接DataFrameに入れるとValueErrorになる。今回はアヤメのデータをロードしたがデータを付け足す場合はPandasのSeries等にindexを付加し結合させていく。その時にデータの欠損などがありnanと入力されたりする事がありそれをどう処理するかが前処理であるデータクレンジングの要になる。例えば０を入れたりラベルに対応する平均を入れたりする。

```python
df =pd.concat([df0, df1], axis=1)
df
```

# 【課題4】irisデータセット自体の調査

Iris属 (アヤメ属) はアヤメ科に含まれる属のうちのひとつで，世界の温帯に150種が知られているようです (アヤメ属 - Wikipedia)．irisデータセットに含まれる
Iris setosa，Iris versicolor，Iris virginicaは，このアヤメ属に含まれる植物のうちの3種

論文の骨子
アヤメ属の3種 (Iris setosa，Iris versicolor，Iris virginica) の種分化について，形態，遺伝，生態，進化の考察を行なっている．

setosaに比べて，versicolorとvirginicaの形態はよく似ている．
setosaとの違いは，virsinicaのほうが大きく，versicolorで小さい．
versicolorは，setosaとvirginicaの雑種に由来しているかもしれない．

形態
全体的な形態はよく似ているが，sepal (がく片) とpetal (花弁) の長さ・幅が，特に種間で異なる特徴である
[参考サイト](http://d.hatena.ne.jp/tsutatsutatsuta/20121206/1354737461)

.locや.iloc等。この２つはどう違うのか  
locは行ラベル、 列ラベルで指定する　df.loc[50:99,['petal length (cm)']]　等  
ilocは行の番号(0 ~ )、列の番号(0 ~ )で指定する　df.iloc[:,[1]]等  
ixは行ラベル、列ラベル または 行の番号(0 ~ )、列の番号(0 ~ )で指定する

```python
from sklearn import datasets
iris = datasets.load_iris()
features = iris.data
feature_names = iris.feature_names
correlation_matrix = np.corrcoef(features.transpose())
# 相関行列のヒートマップを描く
sns.heatmap(correlation_matrix, annot=True,
            xticklabels=feature_names,
            yticklabels=feature_names)
plt.show
```

# 【課題8】簡単な考察

散布図を見るとsetosaと(versicolor,virginica)のくくりで大きく分布が異なっているのが分かる。特にpetalが特徴的でsetosaが大きな特徴を示している。
setosaはpetalが小さい傾向にあるのが散布図から解り。versicolor,virginicaは肥大傾向にある。特にvirginicaは全体的に大きい。
これによりsetosaとversicolor,virginicaは進化の過程で割と早く分科しているものと思われる。
