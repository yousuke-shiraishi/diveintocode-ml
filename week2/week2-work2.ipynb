# 【課題2】2次元配列を作ろう

```python
X = np.linspace(0,1.0,101)
y = X*0.5 +1
plt.xlim(0,1)
plt.ylim(0,2)
plt.plot(X,y)
plt.show()
```

# 【課題3】勾配を求めよう

```python
X = np.linspace(0,1.0,101)
y = X*0.5 +1
xy_graph = np.concatenate([X,y])
plt.xlim(0,1)
plt.ylim(0,2)
plt.plot(X,y)

X2 = np.linspace(0.01,1.01,101)
y2 = X2*0.5+1
print((y2 - y)/ (X2 - X))
```

# 【課題4】配列、勾配、グラフを作ろう

```python
X = np.linspace(0,1.0,101)
y = X**2
xy_graph = np.concatenate([X,y])

X2 = np.linspace(0.01,1.01,101)
y2 = X2**2
print((y2 - y)/ (X2 - X))
plt.xlim(0,1)
plt.ylim(0,2)
plt.plot(X,y)
```

# 【課題5】最小値の求め方を説明しよう

```python
X = np.linspace(0,1.0,101)
y = X**2
xy_graph = np.concatenate([X,y])

X2 = np.linspace(0.01,1.01,101)
y2 = X2**2
grad_all = (y2 - y)/ (X2 - X)
#勾配を一つずつ移動、勾配の符号が変化した時かグラフの末端になるまで。
for k in range(len(grad_all)+1):
    if  grad_all[len(grad_all) -k-1] < grad_all[len(grad_all) -(k+2)] or len(grad_all) -(k+2) ==0:
        break
print(grad_all[len(grad_all) -(k+2)])
        
```
