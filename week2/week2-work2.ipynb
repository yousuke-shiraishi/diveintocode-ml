# 【課題4】配列、勾配、グラフを作ろう

```python
X = np.linspace(0,1.0,101)
y = X**2
X2 = np.linspace(0.01,1.01,101)
y2 = X2**2
y3 = y2 - y
X3 = X2 - X
result = y3/X3
plt.xlim(0,1)
plt.ylim(0,2)
#勾配の増加量は線形で増えていくので一直線になる
plt.plot(X,result)
#２次元配列を作る
d2array = np.array([X,result])
print(d2array)
```

# 【課題5】最小値の求め方を説明しよう

勾配降下法とは与えられた数式の最小誤差を求めるために使われる。$y=x^2$が誤差の値だとすればyが最も少ないところが最小二乗誤差の最適値となる。  
というわけでまず始点を決定する。そこから勾配を求め勾配がゼロになるまで降っていく。勾配を求める関数は次に移動する勾配の値を保有するXの位置を求める関数となる。

求められる勾配の関数とは何か？

init_locは現在のXの位置  
next_locは次に移動するXの位置  
gradientはY2-Y1/X2-X1で求められる勾配（今回は配列とする）  

def gradient_descent(init_loc, a):  
&emsp;&emsp;next_loc = int(round(init_loc - a * gradient[init_loc]))  
&emsp;&emsp;return next_loc  

現在のXの値からその勾配gradient[init_loc]に学習率aをかけたものを引くそれが次のXの値になる。それをnext_loc
とするとXの位置next_locの勾配を求める。それが次のXの勾配gradient[next_loc]となる。
